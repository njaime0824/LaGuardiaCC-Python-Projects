```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading CSV file
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#View first 5 entries
df.head()
```

```python
#View last 5 entries
df.tail()
```

```python
#What columns exists and what their data types are
df.info()
```

```python
#What type of Regents Exams are being used within this dataset
df.groupby('Regents Exam').count()
```

```python
#Specifying the Regents Exam to 'Geometry'
df_geometry = df.loc[df['Regents Exam'] == 'Geometry']
```

```python
#Specifying the year to '2016' 
df_geometry = df_geometry.loc[df['Year'] == 2016]
#Renaming 'Year' to 'Year: 2016' and 'Regents Exam' to 'Regents Exam: Geometry
df_geometry_2016 = df_geometry.rename(columns={'Year': 'Year: 2016', 'Regents Exam': 'Regents Exam: Geometry'})
df_geometry_2016
```

```python
#Checking for null values
df_geometry_2016.isnull().sum()
```

```python
#'s' and 'na' are indicative of missing values
#Rename these as missing values in order to work with schools that have non-null values.
df_geometry_2016 = df_geometry_2016.replace({'s': np.nan, 'na': np.nan})
```

```python
#Rechecking for null values
df_geometry_2016.isnull().sum()
```

```python
#Not enough info to input values for nulls
#Dropping rows that have null values to work on schools with complete data
df_geometry_2016 = df_geometry_2016.dropna()
df_geometry_2016
```

```python
#DBN refers to District/Borough/Number
#Using an income map of NYC, I will compare schools within District 24 (Corona, Queens; Low-income community) to District 2 (Lower to Midtown Manhattan; High-income community)
#List of schools within District 2 in Manhattan using fuzzy matching
df_district_2 = df_geometry_2016[df_geometry_2016['School DBN'].str.contains('02M')]
df_district_2
```

```python
#List of schools within District 24 in Queens using fuzzy matching
df_district_24 = df_geometry_2016[df_geometry_2016['School DBN'].str.contains('24Q')]
df_district_24
```

```python
#Drop schools that repeat
df_district_2 = df_district_2.drop_duplicates(subset='School Name', keep="first")
df_district_2
```

```python
df_district_24 = df_district_24.drop_duplicates(subset='School Name', keep="first")
df_district_24
```

```python
#Using all the schools that remain from both districts 
#Create a DataFrame consisting of 'School DBN' and 'Mean Score' for analysis
df_district_2 = df_district_2[['School DBN', 'Mean Score']]
df_district_24 = df_district_24[['School DBN', 'Mean Score']]
```

```python
#Convert 'Mean Score' from string to float in order to manipulate
df_district_2['Mean Score'] = df_district_2['Mean Score'].astype(float)
df_district_24['Mean Score'] = df_district_24['Mean Score'].astype(float)

#Keep the 5 schools from each district that have the highest mean scores, i.e, the 5 highest performing schools within both districts
df_district_2 = df_district_2.nlargest(5, 'Mean Score')
df_district_2
```

```python
df_district_24 = df_district_24.nlargest(5, 'Mean Score')
df_district_24
```

```python
#Find the average of the 5 mean scores
df_district_2.mean()
```

```python
df_district_24.mean()
```

The average of the mean scores (from the 5 schools in each district) suggests that schools within District 2 perform better on Geometry Regents Exams. Moreover, it suggests that the average student in District 2 passes the Geometry Regents Exam with a score above 65% (73.4%). While those in District 24, on average, score below 65% (56.88%). This average score can be affected by various factors, income of area, quality of school, level of mathematical literacy of students within the 2016 year, etc. With more data and continued research, we can make further conclusions about student performance within these two districts.
